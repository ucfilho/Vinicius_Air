```python
import numpy as np
from sklearn import linear_model
import statsmodels.api as sm
from numpy import log10 as log10
```

```python
dep_2 = np.array([0.01636, 0.01609, 0.00675, 0.00682, 0.00179, 0.00160, 0.00165, 0.00053, 0.00054])
N = np.array([168, 200, 1000, 1180, 4730, 8035, 5254, 28617, 32650])
log_dep_2 =log10(dep_2)
log_N = log10(N)
x = log_N.reshape(-1,1) 

y = log_dep_2.reshape(-1,1)
```

```python
model = linear_model.LinearRegression()
model.fit(y,x )
print(' A= ',model.intercept_ ,'B =',model.coef_ )
print('r^2',model.score(x, y))
```
